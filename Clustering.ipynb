#CLUSTERING

1.  What is unsupervised learning in the context of machine learning ?

- Unsupervised learning is a type of machine learning that uses algorithms to find patterns and structures in unlabeled data without human intervention. Instead of learning from pre-defined inputs and outputs like in supervised learning, the goal is for the model to discover hidden relationships, group similar data points, or reduce data complexity on its own. This technique is useful for tasks like exploratory data analysis, clustering, and anomaly detection.

2. How does K-Means clustering algorithm work ?

- K-means is an iterative, centroid-based clustering algorithm that partitions a dataset into similar groups based on the distance between their centroids. The centroid, or cluster center, is either the mean or median of all the points within the cluster depending on the characteristics of the data.

3. Explain the concept of a dendrogram in hierarchical clustering.

- A dendrogram is a tree-like diagram used in hierarchical clustering to visualize the relationships between data points, showing how clusters are merged or divided at each step. It displays a hierarchy of clusters, where the height of each merge point indicates the distance or dissimilarity between the clusters being joined; lower merges represent more similar clusters.

4. What is the main difference between K-Means and Hierarchical Clustering ?

- The main difference between K-Means and Hierarchical Clustering is that K-Means is a partitional clustering method that requires the number of clusters (K) to be specified in advance, while Hierarchical Clustering builds a tree-like structure of clusters (a dendrogram) without needing a pre-specified number of clusters.

5. What are the advantages of DBSCAN over K-Means ?

- DBSCAN's advantages over K-Means include its ability to discover arbitrarily-shaped clusters and its automatic identification of noise/outlier points without prior knowledge of the number of clusters. Unlike K-Means, which assumes clusters are spherical and requires the number of clusters (k)to be specified beforehand, DBSCAN identifies clusters based on data density.

6. When would you use Silhouette Score in clustering ?

- You would use Silhouette Score to evaluate the quality of clusters by measuring how well each data point fits into its own cluster compared to other clusters. It is especially useful for determining the optimal number of clusters (k) in algorithms like K-Means, as the highest silhouette score indicates the most appropriate number of clusters. It can also be used to compare different clustering results to see which one makes the most sense for a given dataset.

7. What are the limitations of Hierarchical Clustering ?

- The main disadvantages of hierarchical clustering are its high computational and space complexity, making it inefficient for large datasets. It is also sensitive to noise and outliers, cannot undo previous merges, and can be difficult to interpret, especially in determining the correct number of clusters from the dendrogram.

8.  Why is feature scaling important in clustering algorithms like K-Means ?

- Feature scaling is crucial in K-Means clustering because the algorithm relies on distance calculations (typically Euclidean distance) to determine the similarity between data points. Without scaling, features with larger numerical ranges or magnitudes will dominate the distance metrics, leading to biased results and suboptimal cluster formation.

9. How does DBSCAN identify noise points ?

- DBSCAN identifies a point as noise if it is neither a core point nor a border point. This effectively labels points that lie in low-density regions as outliers, which are not considered part of any cluster.

10. Define inertia in the context of K-Means .

- In the context of K-Means clustering, inertia, also known as the within-cluster sum of squares (WCSS), is a metric that measures the compactness of the clusters.

Definition

Inertia is calculated as the sum of the squared distances between each data point and the centroid of the cluster to which it has been assigned, across all clusters. The primary objective of the K-Means algorithm is to minimize this total inertia value.

11. What is the elbow method in K-Means clustering ?

- The elbow method is a heuristic used to determine the optimal number of clusters  for a dataset in K-Means clustering . The goal of the method is to find the point where the addition of more clusters no longer significantly improves the modeling of the data .

12. Describe the concept of "density" in DBSCAN .

- In DBSCAN, "density" refers to the concept that data points are clustered together in regions where they are closely packed, separated by areas of low density. The algorithm defines a "dense" region using two parameters: an epsilon distance (\(\varepsilon \)) and a minimum number of points (\(minPts\)). A point is considered "dense" if there are at least \(minPts\) points within its \(\varepsilon \)-neighborhood, and clusters are formed by connecting these dense regions.

13.  Can hierarchical clustering be used on categorical data ?

- Yes, hierarchical clustering can be used on categorical data, provided that an appropriate distance or dissimilarity measure is defined for this type of data. Standard metrics like Euclidean distance are not suitable for categorical variables because they assume a meaningful numerical order and continuous space.

14. What does a negative Silhouette Score indicate ?

- A negative Silhouette Score indicates that a data point has likely been assigned to the wrong cluster. This suggests that the clustering algorithm is performing poorly, as the data point is closer to a different cluster than it is to its own. A negative overall score means that many data points are misclassified, and the clusters are poorly defined and overlapping.

15. Explain the term "linkage criteria" in hierarchical clustering

- In hierarchical clustering, the linkage criteria (or linkage method) is the rule that defines the distance between two clusters. This distance measure is crucial because the algorithm iteratively merges the two closest clusters, and the choice of linkage method significantly influences the shape and size of the resulting clusters.

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities ?

- K-Means clustering performs poorly on data with varying cluster sizes or densities because the algorithm inherently assumes that clusters are spherical, equally sized, and have similar densities.

17. What are the core parameters in DBSCAN, and how do they influence clustering ?

- The core parameters in the DBSCAN algorithm are Epsilon (Îµ) and Minimum Points (MinPts). These parameters are fundamental to the algorithm's concept of density and significantly influence the clustering results, including the number, size, and shape of the clusters detected.

18.  How does K-Means++ improve upon standard K-Means initialization ?

- K-Means++ improves upon standard K-Means initialization by employing a systematic and probabilistic method to ensure that initial centroids are well-distributed and far apart across the dataset, rather than selecting them completely at random.

19. What is agglomerative clustering ?

- Agglomerative clustering is a "bottom-up" hierarchical clustering algorithm that groups data points into clusters based on their similarity. It begins by treating each individual data point as a separate cluster and then iteratively merges the closest pairs of clusters until all data points belong to a single, large cluster.

20. What makes Silhouette Score a better metric than just inertia for model evaluation?

- Silhouette score is a better metric than inertia because it considers both cluster compactness (how close points are to their own cluster's centroid) and separation (how far apart different clusters are), while inertia only measures the within-cluster sum of squares. A low inertia value can be achieved by simply increasing the number of clusters, but this may not result in well-separated, meaningful groups. The silhouette score provides a more holistic view of clustering quality, with higher scores indicating better-defined clusters.



```python
#21.  Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import numpy as np


X, y_true = make_blobs(n_samples=300, centers=4, n_features=2,
                       cluster_std=0.60, random_state=0)


kmeans = KMeans(n_clusters=4, init='k-means++', n_init=10, random_state=0)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)


centers = kmeans.cluster_centers_


plt.figure(figsize=(10, 6))


plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', label='Data points')


plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', edgecolors='black', label='Centroids')

plt.title('K-Means Clustering with 4 Centers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#23. Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot

import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np


X, y_true = make_moons(n_samples=300, noise=0.1, random_state=42)


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


dbscan = DBSCAN(eps=0.3, min_samples=5)
clusters = dbscan.fit_predict(X_scaled)


plt.figure(figsize=(8, 6))


unique_labels = set(clusters)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]


for k, col in zip(unique_labels, colors):
    if k == -1:

        col = 'k'
        marker = 'x'
    else:
        marker = 'o'


    class_member_mask = (clusters == k)
    xy = X_scaled[class_member_mask]


    plt.scatter(xy[:, 0], xy[:, 1], color=col, marker=marker, s=50, label=f'Cluster {k}' if k!=-1 else 'Outlier/Noise')

plt.title('DBSCAN Clustering with Outliers Highlighted')
plt.xlabel('Scaled Feature 1')
plt.ylabel('Scaled Feature 2')
plt.legend()
plt.show()
```

```python
#24. Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each cluster

import pandas as pd
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np


wine = datasets.load_wine()
X = wine.data
y = wine.target

print(f"Original dataset shape: {X.shape}")


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42, n_init=10)
kmeans.fit(X_scaled)
cluster_labels = kmeans.labels_


cluster_sizes = np.bincount(cluster_labels)

print("\nCluster sizes (number of samples in each cluster):")
for i, size in enumerate(cluster_sizes):
    print(f"Cluster {i}: {size} samples")


actual_class_sizes = np.bincount(y)
print("\nActual class distribution in the dataset:")
for i, size in enumerate(actual_class_sizes):
    print(f"Class {i}: {size} samples")
```

```python
#25.  Use make_circles to generate synthetic data and cluster it using DBSCAN. Plot the result

import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
import numpy as np


X, y = make_circles(n_samples=750, factor=0.3, noise=0.1, random_state=42)


dbscan = DBSCAN(eps=0.15, min_samples=5)
clusters = dbscan.fit_predict(X)


plt.figure(figsize=(8, 6))


unique_labels = set(clusters)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

# Create a boolean mask for core samples
is_core = np.zeros_like(clusters, dtype=bool)
is_core[dbscan.core_sample_indices_] = True

for k, col in zip(unique_labels, colors):
    if k == -1:
        col = [0, 0, 0, 1] # Black for noise points

    class_member_mask = (clusters == k)

    # Plot core points for the current cluster
    xy_core = X[class_member_mask & is_core]
    plt.plot(xy_core[:, 0], xy_core[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=8, label=f'Cluster {k} Core' if k != -1 else 'Noise')

    # Plot border points for the current cluster
    xy_border = X[class_member_mask & ~is_core]
    plt.plot(xy_border[:, 0], xy_border[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=4, label=f'Cluster {k} Border' if k != -1 else '')

plt.title('DBSCAN Clustering on make_circles data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
#29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


n_samples = 300
n_features = 2
true_centers = 3
X, y_true = make_blobs(n_samples=n_samples, n_features=n_features, centers=true_centers,
                       cluster_std=1.0, random_state=42)


k_values = range(2, 6)


silhouette_scores = {}


for k in k_values:

    kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, random_state=42)
    cluster_labels = kmeans.fit_predict(X)


    score = silhouette_score(X, cluster_labels)
    silhouette_scores[k] = score
    print(f"For k = {k}, the average silhouette score is: {score:.4f}")

plt.figure(figsize=(8, 5))
plt.bar(list(silhouette_scores.keys()), list(silhouette_scores.values()), color='skyblue')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Average Silhouette Score")
plt.title("Silhouette Scores for Different Numbers of Clusters")
plt.xticks(list(silhouette_scores.keys()))
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
#30. Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage

import pandas as pd
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

linked = linkage(iris_df, method='average')

plt.figure(figsize=(10, 7))
dendrogram(linked,
           orientation='top',
           labels=iris.target_names[iris.target],
           distance_sort='descending',
           show_leaf_counts=True)
plt.title('Dendrogram of Iris Dataset (Average Linkage)')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
#31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from matplotlib.colors import ListedColormap


X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=2.0, random_state=42)


kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)


x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))


Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)


cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])


plt.figure(figsize=(10, 7))
plt.contourf(xx, yy, Z, cmap=cmap_light, alpha=0.6)


plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap=cmap_bold,
            s=50, edgecolor='k')


centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], marker='X', c='red', s=200, alpha=1, label='Centroids')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title("K-Means Clustering with Overlapping Clusters and Decision Boundaries")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
#32.  Load the Digits dataset and apply DBSCAN after reducing dimensions with t-SNE. Visualize the results

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


digits = datasets.load_digits()
X = digits.data
y_true = digits.target
n_samples, n_features = X.shape
print(f"Dataset shape: {X.shape}")


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


pca = PCA(n_components=50)
X_pca = pca.fit_transform(X_scaled)


tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, random_state=42)
X_tsne = tsne.fit_transform(X_pca)
print(f"Reduced data shape: {X_tsne.shape}")


dbscan = DBSCAN(eps=3, min_samples=5)
clusters = dbscan.fit_predict(X_tsne)
labels = dbscan.labels_

n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)
print(f"Estimated number of clusters: {n_clusters_}")
print(f"Estimated number of noise points: {n_noise_}")


plt.figure(figsize=(8, 6))

scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='viridis', s=20, alpha=0.8)

plt.colorbar(scatter, ticks=range(-1, n_clusters_))
plt.title('DBSCAN Clustering on Digits Dataset (t-SNE Reduced)')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.axis('off')
plt.show()
```

```python
#40. Load the Digits dataset, apply PCA to reduce to 3 components, then use KMeans and visualize with a 3D scatter plot.

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

digits = datasets.load_digits()
X = digits.data
y = digits.target


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)


kmeans = KMeans(n_clusters=10, init='k-means++', random_state=42, n_init=10)
kmeans.fit(X_pca)

labels = kmeans.labels_


centroids = kmeans.cluster_centers_

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=labels, cmap='viridis', s=50, alpha=0.8)


ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:, 2], marker='x', s=100, color='black', label='Centroids')

ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('KMeans Clustering of Digits Dataset (PCA-reduced to 3D)')
ax.legend()

plt.show()
```

```python
#41. Generate synthetic blobs with 5 centers and apply KMeans. Then use silhouette_score to evaluate the clustering

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np


n_samples = 300
n_features = 2
n_clusters_true = 5
random_state = 42

X, y_true = make_blobs(n_samples=n_samples, n_features=n_features,
                       centers=n_clusters_true, cluster_std=0.8,
                       random_state=random_state)

print(f"Generated data shape: {X.shape}")


n_clusters_kmeans = 5
kmeans = KMeans(n_clusters=n_clusters_kmeans, random_state=random_state, n_init=10)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print(f"KMeans labels generated: {len(labels)} samples")


silhouette_avg = silhouette_score(X, labels)
print(f"The average silhouette score for {n_clusters_kmeans} clusters is: {silhouette_avg:.4f}")


plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.8)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title(f'KMeans Clustering with {n_clusters_kmeans} Clusters (Silhouette Score: {silhouette_avg:.4f})')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#42. Load the Breast Cancer dataset, reduce dimensionality using PCA, and apply Agglomerative Clustering. Visualize in 2D

import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
import numpy as np


cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
target_names = cancer.target_names


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)


pca_df = pd.DataFrame(data=X_pca, columns=['principal component 1', 'principal component 2'])


agg_clustering = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
cluster_labels = agg_clustering.fit_predict(X_pca)


pca_df['cluster_label'] = cluster_labels


plt.figure(figsize=(10, 7))


colors = ['red', 'green']

for cluster_label, color in zip(pca_df['cluster_label'].unique(), colors):
    plt.scatter(pca_df[pca_df['cluster_label'] == cluster_label]['principal component 1'],
                pca_df[pca_df['cluster_label'] == cluster_label]['principal component 2'],
                c=color, label=f'Cluster {cluster_label}', s=50)

plt.title('Agglomerative Clustering Results (PCA Reduced to 2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#43. Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN side-by-side

import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np


X, y_true = make_circles(n_samples=500, noise=0.1, factor=0.5, random_state=42)


X = StandardScaler().fit_transform(X)


kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X)


dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)


fig, axes = plt.subplots(1, 3, figsize=(18, 6))


axes[0].scatter(X[:, 0], X[:, 1], c=y_true, cmap='Paired', s=50)
axes[0].set_title("Original Noisy Data")


axes[1].scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='Paired', s=50)
axes[1].set_title(f"KMeans Clustering (k=2)")


cmap = plt.cm.get_cmap('Paired', len(np.unique(dbscan_labels)))
axes[2].scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap=cmap, s=50)
axes[2].set_title(f"DBSCAN Clustering (eps=0.3, min_samples=5)")

plt.tight_layout()
plt.show()
```

```python
#44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score


iris = load_iris()
X = iris.data


n_clusters = 3
clusterer = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
cluster_labels = clusterer.fit_predict(X)


sample_silhouette_values = silhouette_samples(X, cluster_labels)
average_silhouette_score = silhouette_score(X, cluster_labels)
print(f"The average silhouette score is: {average_silhouette_score:.3f}")


fig, ax1 = plt.subplots(1, 1)
fig.set_size_inches(9, 7)


ax1.set_xlim([-0.1, 1])

ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

y_lower = 10
for i in range(n_clusters):

    ith_cluster_silhouette_values = \
        sample_silhouette_values[cluster_labels == i]

    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    color = cm.nipy_spectral(float(i) / n_clusters)
    ax1.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_values,
                      facecolor=color, edgecolor=color, alpha=0.7)


    ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))


    y_lower = y_upper + 10

ax1.set_title("Silhouette Plot for the Iris Dataset (KMeans, n_clusters=3)")
ax1.set_xlabel("Silhouette coefficient values")
ax1.set_ylabel("Cluster label")


ax1.axvline(x=average_silhouette_score, color="red", linestyle="--", label="Average Silhouette Score")
ax1.legend()

plt.show()
```

```python
#47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count

import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt


X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.8, random_state=42)


dbscan = DBSCAN(eps=0.5, min_samples=8)
labels = dbscan.fit_predict(X)


noise_points_count = np.sum(labels == -1)
print(f"Number of noise points: {noise_points_count}")

unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:

        col = [0, 0, 0, 1]

    class_member_mask = (labels == k)

    xy = X[class_member_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6)

plt.title('DBSCAN clustering of noisy blobs')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
#48.Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


digits = load_digits()
X = digits.data
y = digits.target

print(f"Original data shape: {X.shape}")


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


pca = PCA(n_components=50)
X_pca = pca.fit_transform(X_scaled)
print(f"Data shape after PCA: {X_pca.shape}")


tsne = TSNE(n_components=2, init='pca', random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X_pca)

print(f"Data shape after t-SNE: {X_tsne.shape}")


n_clusters = 10
agglomerative = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
cluster_labels = agglomerative.fit_predict(X_tsne)


plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=cluster_labels, cmap='Spectral', s=10)
plt.legend(handles=scatter.legend_elements()[0], labels=range(n_clusters), title="Cluster")
plt.title('Agglomerative Clustering on t-SNE Reduced Digits Dataset')
plt.xlabel('t-SNE component 1')
plt.ylabel('t-SNE component 2')
plt.axis('off')
plt.show()
```
